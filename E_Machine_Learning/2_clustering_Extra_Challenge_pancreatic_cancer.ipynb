# Clustering on Pancreatic cancer dataset

The pancreatic cancer dataset includes samples collected from multiple sites and patient cohorts. Before carrying out downstream analyses, it is important to assess whether the biomarker measurements show systematic differences related to collection site or cohort membership. One way to explore this is through clustering analysis, which groups samples based on similarity in their measured features without using known class labels. By examining the resulting clusters, we can evaluate whether samples tend to group according to site, cohort, or other technical factors, and whether these sources of variation may influence later analyses.

The patient cohorts can be found from `patient_cohort` column and collection site information in `sample_origin` column. We will perform clustering with and without these to see how having them in the analysis affects things.
